```python
import pandas as pd
import numpy as np
import keras
import os
```

```python
from keras.applications.xception import Xception, preprocess_input
from keras.preprocessing.image import load_img, img_to_array
import glob
```

```python
data_dir = 'C:/Users/tsharmili/Downloads/cnn/waffle_pancakes'
```

```python
def preprocess(im):
    im = img_to_array(im)
    im = np.expand_dims(im , axis=0)
    im = preprocess_input(im)
    return im[0]
```

```python
os.chdir(os.path.join(data_dir,'train','waffles'))
```

```python
waffle_train_img_names = glob.glob('*.png')
```

```python
waffle_train =[]
for i in waffle_train_img_names:
    im = load_img(i, target_size=(224,224,3))
    waffle_train.append(preprocess(im))
```

```python
#waffle_train
```

```python
os.chdir(os.path.join(data_dir,'test','waffles'))
waffle_test_img_names = glob.glob('*.png')
waffle_test =[]
for i in waffle_test_img_names:
    im = load_img(i, target_size=(224,224,3))
    waffle_test.append(preprocess(im))
```

```python
os.chdir(os.path.join(data_dir,'train','pancakes'))
pancakes_train_img_names = glob.glob('*.png')
pancakes_train =[]
for i in pancakes_train_img_names:
    im = load_img(i, target_size=(224,224,3))
    pancakes_train.append(preprocess(im))
```

```python
os.chdir(os.path.join(data_dir,'test','pancakes'))
pancakes_test_img_names = glob.glob('*.png')
pancakes_test =[]
for i in pancakes_test_img_names:
    im = load_img(i, target_size=(224,224,3))
    pancakes_test.append(preprocess(im))
```

```python
waffle_train = np.array(waffle_train,dtype = 'float32')
waffle_test = np.array(waffle_test,dtype = 'float32')
pancakes_train = np.array(pancakes_train,dtype = 'float32')
pancakes_test = np.array(pancakes_test,dtype = 'float32')
```

```python
train = np.concatenate((waffle_train,pancakes_train),axis=0)
```

```python
train.shape
```

```python
y = [0]*waffle_train.shape[0] + [1]*pancakes_train.shape[0]
y
```

```python
y = np.array(y,dtype='float32')
```

```python
model = Xception(include_top=False, pooling ='avg',input_shape=(224,224,3))
```

```python
train_features = model.predict(train)
```

```python
train_features.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf= LogisticRegression()
```

```python
clf.fit(train_features,y)
```

```python
test = np.concatenate((waffle_test,pancakes_test),axis=0)
```

```python
from keras.applications.inception_v3 import InceptionV3, preprocess_input
```

```python
data_dir
```

```python
def preprocess(x):
    X = np.expand_dims(x,axis=0)
    X= preprocess_input(X)
    return X[0]

data_gen = image.ImageDataGenerator(rotation_range=40,
                                   shear_range=0.2,
                                   horizontal_flip=True,
                                   vertical_flip=False,
                                   zoom_range=0.2,
                                   fill_mode='nearest',
                                   preprocessing_function=preprocess)

train_generator = data_gen.flow_from_directory(os.path.join(data_dir,'train'),
                                              target_size=(150,150))

valid_generator = data_gen.flow_from_directory(os.path.join(data_dir,'test'),
                                              target_size=(150,150))
```

```python
model = InceptionV3(include_top=False, pooling ='avg',input_shape=(150,150,3),
                weights= 'imagenet')
```

```python
model.summary()
```

```python
from keras.layers import Dense
from keras.models import Model

x = model.output
x = Dense(1024, activation='relu')(x)
predictions = Dense(2, activation ='softmax')(x)

model1 = Model(inputs= model.input , outputs= predictions)
```

```python
model1.summary()
```

```python
for layers in model.layers:
    layers.trainable=False
```

```python
model1.compile(loss = 'categorical_crossentropy',
             optimizer='rmsprop',
             metrics=['accuracy'])
```

```python
batch_size=32
```

```python
model1.fit_generator(train_generator,
                    validation_data=valid_generator,
                    epochs=3,
                    verbose=1,shuffle=False,
                    steps_per_epoch= 735//batch_size,
                    validation_steps=389//batch_size)
```
