```python
import os
import tensorflow as tf
import keras
```

```python
#importing models

from keras.applications.vgg16 import VGG16, preprocess_input

#preproceesing is to be done on the model
#load_img is for loading
#img to array is for converting to array

from keras.preprocessing.image import load_img, img_to_array
import numpy as np
```

```python
model = VGG16()
```

```python
model.summary()
```

```python
#img_path = 'C:/Users/tsharmili/Downloads/cnn/pizza.jpg'
img= load_img('pizza.jpg', target_size=(224,224))
```

```python
img.size
```

```python
img_array = img_to_array(img)
img_array.shape
```

```python
#preprocess_input takes 4d input
#so we have to add a new dimension

img_array = np.expand_dims(img_array,axis=0)
```

```python
img_array.shape
#this is what we pass to the vgg16 model
```

```python
img_pre_processed = preprocess_input(img_array)
img_pre_processed
```

```python
from keras.applications.vgg16 import decode_predictions
```

```python
preds= model.predict(img_pre_processed)
```

```python
preds
```

```python
decode_predictions(preds,top=5)
#our model is able to identify the image with accuracy of 99%
```
